```python
#3.2: create a random 10x10 array named A
randomMatrix = np.random.rand(10,10)
randomMatrix
```

```python
#3.3: return only the odd rows of A
oddRows = randomMatrix[::2]
oddRows
```

```python
#3.4: perform a matrix multiplication of A with it's transposed
multiMatrix = np.dot(randomMatrix, np.transpose(randomMatrix))
multiMatrix
#multiMatrix2 = np.matrix(randomMatrix) * np.matrix(randomMatrix.T)
#multiMatrix2
```

```python
#3.5: compute the matrix inverse of A
inverseMatrix = np.matrix(randomMatrix)* np.linalg.inv(randomMatrix)
inverseMatrix
#inverseMatrix2 = np.matrix(randomMatrix)* np.matrix(randomMatrix).I
#inverseMatrix2
```
